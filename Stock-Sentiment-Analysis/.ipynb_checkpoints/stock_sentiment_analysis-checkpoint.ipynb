```python
import pandas as pd
```

```python
df = pd.read_csv("Data.csv",encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df["Date"]
```

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
# Remove punctuations
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]",' ',regex=True, inplace=True)
```

```python
# Renaming column names for easiness to access

list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index

data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
data.iloc[0,0:25]
```

```python
headlines = []

for i in range(0,len(data)) :
    headlines.append(' '.join(sen for sen in data.iloc[i,0:25] ))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
traindataset.shape, train.shape
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_set= []
for row in range(0,len(test.index)):
    test_set.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_set)
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
